Separates login hours into even-hour logins and odd-hour logins.

```python
login_hours = [1, 3, 5, 7, 2, 6, 8, 4, 9, 10]
odd_hrs = []
even_hrs = []
for i in range(len(login_hours)):
    if login_hours[i]%2 == 0:
        even_hrs.append(login_hours[i])
    else:
        odd_hrs.append(login_hours[i])

print(f"Actual login hours :: {login_hours}")
print(f"Even login hours :: {even_hrs}")
print(f"Odd login hours :: { odd_hrs}")
```

Identifies login hours that fall into peak activity time (hours greater than 5).
Calculates how many logins occurred during peak hours.

```python
login_hours = [1, 3, 5, 7, 2, 6, 8, 4, 9, 10]
pick_hrs = []
for i in range(len(login_hours)):
    if login_hours[i] > 5:
        pick_hrs.append(login_hours[i])

print(f"Pick hrs :: {pick_hrs}")
print(f"Pick hrs count :: {len(pick_hrs)}")
        
```
