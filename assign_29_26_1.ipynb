String to Integer Conversion

```python
conv_data = ["10", "20", "abc", "30"]
valid_conv_data = []
for value in conv_data:
    try:
        valid_conv_data.append(int(value))
    except Exception as e:
        print(f"got error for {e}")

print(valid_conv_data)
```

Validate Phone Numbers Extract valid 10-digit phone numbers from: text = "Call 9876543210 or 12345"

```python
text = "Call 9876543210 or 12345"

phone_nos = re.findall("\d+",text)

for phone in phone_nos:
    if re.fullmatch("\d{10}", phone):
        print(f"phone {phone} is valid")
    else:
        print(f"phone {phone} not valid")
```

Extract Dates Extract date from: text = "Today is 2026-03-28" Return: Year Month Day

```python
text = "Today is 2026-03-28"

match = re.search("(\d{4})-(\d{2})-(\d{2})",text)

if match:
    year, month, day = match.groups()
print(f"year {year}")
print(f"month {month}")
print(f"day {day}")
```

Extract numbers followed by "rs": text = "100rs 200usd 300rs"

Extract numbers that come after "rs": text = "rs100 rs200 usd300"

```python
text = "rs100 rs200 usd300"
value_rs = re.findall("rs(\d+)",text)
print(value_rs)
```
