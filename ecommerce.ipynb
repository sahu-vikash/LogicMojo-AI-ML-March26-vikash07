Combines the product names and their prices into a single data structure representing a table.

```python
products = ["Laptop","Phone","Tablet","Monitor","Headphones"] 
amazon_prices = [75000, 45000, 30000, 20000, 5000] 
flipkart_prices = [72000, 47000, 31000, 19500, 4800]

print(f"{'Product ':<10} | {'amazon_prices':<13} | {'flipkart_prices':<15} | {'Difference' :<15}")
print("-" * 55)

combined = zip(products,amazon_prices,flipkart_prices)
for row in combined:
    print(f"{row[0]:<10} | {row[1]:<13} | {row[2]:<15} | {abs(row[1] - row[2]):<15}")   
```

Compares the price of each product across both platforms.

Displays the product name, cheaper platform, and price difference.

```python
print(f"{'product name ':<15} | {'cheaper platform':<15} | {'price difference.':<15}")
print("-" * 55)
combined = zip(products,amazon_prices,flipkart_prices)
for row in combined:
    vendor=''
    if row[1]-row[2] > 0:
        vendor = 'FlipKart'
    else:
        vendor = 'Amazon'
    print(f"{row[0]:<15} | {vendor:<15} | {abs(row[1] - row[2]):<15}")   
```
