***In optimization landscapes of neural networks, the concepts of local minima and global minima are extremely important. Which statement correctly differentiates them?***

A. Local minsima is always the absolute smallest value everywhere

B. Global minima exists only for convex functions

**C. Local minima is smallest in a nearby region, while global minima is smallest overall** ***Correct***

D. Local minima and global minima are always identical

### Explanation

As part of the graph of Gradient Descent optimization:
* **Blue Line**: Represents the loss function.
* **Scattered Red Dots**: Represent the gradient descent steps as it moves from $-1$ towards $0$, which is a local minimum. 

Looking further at the graph, more minima points are available:

* **A.** It classifies that a local minimum is not the absolute smallest value. **[False]**
* **B.** If we check the graph, the smallest minimum is around $-8$ (global minimum). Since the available graph is not convex, it makes the statement. **[False]**
* **C.** Multiple minimum points are available and the first minimum is the smallest in the nearby region, while the global minimum is the absolute smallest (around $-8$). **[True]**
* **D.** Only in the case of a convex function will the local and global minima be identical. Because this graph is not convex, they are different. **[False]**


```python
#function declaration
def f(x):
    return x**4 - 3*x**3

#first differentiation
def df(x):
    return 4*x**3 - 9*x**2


#define a gradient descent for first local minima
starting_point=-1
learning_rate = 0.01
steps = 50

#initializing list to add all the variation point
starting_point_cont = [starting_point]

for i in range(steps):
    #Gradient descent update x = x - learning_rate * f'(x)
    starting_point = starting_point - learning_rate * df(starting_point)
    starting_point_cont.append(starting_point)


#define a gradient descent for Global absolute minima
starting_point_second_minima=1

#initializing list to add all the variation point
starting_point_cont_2 = [starting_point_second_minima]

for i in range(steps):
    #Gradient descent update x = x - learning_rate * f'(x)
    starting_point_second_minima = starting_point_second_minima - learning_rate * df(starting_point_second_minima)
    starting_point_cont_2.append(starting_point_second_minima)

x_range = np.linspace(-2,3, 100)
y_range = f(x_range)

plt.plot(x_range, y_range,label = 'loss function')
plt.scatter(starting_point_cont, [f(step) for step in starting_point_cont], color='red', label='Gradient descent steps')
plt.scatter(starting_point_cont_2, [f(step) for step in starting_point_cont_2], color='skyblue', label='Gradient descent steps')
plt.title("Gradient Descent optimization")
plt.xlabel("x")
plt.ylabel("loss function f(x)")

# Arrow for local minima
plt.annotate('1 Local Minima', 
             xy=(starting_point_cont[len(starting_point_cont)-1], f(starting_point_cont[len(starting_point_cont)-1])),
             xytext=(starting_point_cont[len(starting_point_cont)-1] - 0.5, f(starting_point_cont[len(starting_point_cont)-1]) + 8),
             arrowprops=dict(facecolor='black', shrink=0.05, width=1, headwidth=6))

plt.annotate('Global absolute minima', 
             xy=(starting_point_cont_2[len(starting_point_cont_2)-1], f(starting_point_cont_2[len(starting_point_cont_2)-1])),
             xytext=(starting_point_cont_2[len(starting_point_cont_2)-1] - 0.5, f(starting_point_cont_2[len(starting_point_cont_2)-1]) + 8),
             arrowprops=dict(facecolor='black', shrink=0.05, width=1, headwidth=6))

plt.legend()
plt.show()
```

***In Linear Algebra, eigenvectors are heavily used in PCA, recommendation systems, and computer vision applications. What is the most accurate interpretation of an eigenvector?***
    
A. A vector that completely disappears after transformation

**B. A vector whose direction remains unchanged after transformation** ***Answer***

C. A random vector generated from covariance matrices

D. A vector perpendicular to every other vector 

```python
A = np.array([[2,1],[1,2]])

#Finding eigen value and eigen vector
eig_val, eig_vec = np.linalg.eig(A)

print(f"eigen values {eig_val}")
print(f"eigen vector {eig_vec}")

#extract first eigen value and vector
first_vector = eig_vec[:,0]
first_value  = eig_val[0]

second_vector = eig_vec[:,1]
second_value  = eig_val[1]

#Matrix transformation 1
trns_matrix = A @ first_vector

#Matrix transformation 2
trns_matrix_1 = A @ second_vector

x_val = np.linspace(-3,3,10)
plt.plot(x_val, ((first_vector[1]/first_vector[0])*x_val), color='cyan', alpha=0.2, label='Direction Line first vector')
plt.plot(x_val, ((second_vector[1]/second_vector[0])*x_val), color='cyan', alpha=0.2, label='Direction Line second vector')
plt.quiver(0, 0, first_vector[0], first_vector[1], angles='xy', scale_units='xy', scale=1, 
           color='blue', width=0.01, label=f'Original v1 {first_vector.round(2)}')

plt.quiver(0, 0, trns_matrix[0], trns_matrix[1], angles='xy', scale_units='xy', scale=1, 
           color='blue', width=0.006, label=f'Transformed matrix 1 {trns_matrix.round(2)}')

plt.quiver(0, 0, second_vector[0], second_vector[1], angles='xy', scale_units='xy', scale=1, 
           color='red', width=0.01, label=f'Original v2 {second_vector.round(2)}')

plt.quiver(0, 0, trns_matrix_1[0], trns_matrix_1[1], angles='xy', scale_units='xy', scale=1, 
           color='red', width=0.01, label=f'Transformed matrix 2 {trns_matrix_1.round(2)}')

plt.gca().set_aspect('equal', adjustable='box')

plt.xlim([-3, 3])
plt.ylim([-3, 3])
plt.grid(True, ls=':')
plt.xlabel('X')
plt.ylabel('Y')
plt.title(f'Eigenvector Direction Stays on the Same Line (λ ≈ {first_value:.2f})')
#plt.legend(loc='upper right')
plt.show()
```

***Evaluate the following definite integral frequently used in probability and area-under-curve calculations:
Integral from 0 to 1 of x^2 dx***

A. 1/2

**B. 1/3** ***Answer***

C. 1

D. 2 

__As mentioned below the area under the curve for definitive integral frequency__

```python
#defining function
def f(x):
    return x**2

result, error = integrate.quad(f,0,1)
print(f"definite integral {result}")
print(f"Error {error}")

#Generating area under the curve for above result
x_val = np.linspace(-0.2, 1.5, 100)
y_val = f(x_val)

#for filling (as have limit) area under the curve starting 0 end 1.
x_fill = np.linspace(0,1,100)
y_fill = f(x_fill)

plt.plot(x_val, y_val, color='blue', linewidth=2, label='$f(x) = x^2$')
plt.fill_between(x_fill, y_fill, color='lightblue', alpha=0.5, 
                 label=f'Area = {result:.3f}')

#setting vertical boundry as area start from 0 and end with 1.
plt.vlines([0, 1], 0, [f(0), f(1)], 
           color='red', linestyle='--', alpha=0.7)

plt.xlim([-0.2, 1.4])
plt.ylim([0, 1.4])
plt.grid(True, ls=':', alpha=0.6)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('definite integral frequently')
plt.legend(loc='upper left')
plt.show()
```

Consider the following mathematical functions frequently used in AI/ML:

log(x), x^2, x, e^x

As the value of x becomes extremely large, which function grows at the fastest rate and dominates all others?

A. log(x)

B. x^2

C. x

***D. e^x*** **Answer**

Note: Hirearchy of mathematical function **log(x) < x < x^2 < e^x**

```python
# Generate 20 points between 1 and 5
x = np.linspace(1, 5, 20)

# Calculate the function values
y_log = np.log(x)
y_linear = x
y_square = x**2
y_exp = np.exp(x)

fig, ax1 = plt.subplots(figsize=(14, 6))

# Linear Scale (Shows exponential explosion)
ax1.plot(x, y_log, label='log(x)', color='blue', linewidth=0.5)
ax1.plot(x, y_linear, label='x', color='yellow', linewidth=1)
ax1.plot(x, y_square, label='x^2', color='green', linewidth=2)
ax1.plot(x, y_exp, label='e^x', color='red', linewidth=2.5)

ax1.set_title('Linear Scale: e^x Dominates Completely', fontsize=12)
ax1.set_xlabel('x', fontsize=10)
ax1.set_ylabel('f(x)', fontsize=10)
ax1.grid(True, linestyle='-', alpha=0.5)
ax1.legend(loc='upper left')

plt.tight_layout()
plt.show()
```
