Iterates through the dataset and removes all invalid age values

```python
ages = [25, 42, -1, 34, None, 52, 17, -5, 29, None, 46]
invalid_removed = []
invalid_count = 0
for i,value in enumerate(ages):
    if None is value or value < 0:
        del ages[i]
        invalid_count+=1

invalid_removed.extend(ages)
print(f"FIltered data with removed invalid ages {invalid_removed}")
```

Calculates the average age of valid customers.

```python
total_sum = 0
for age in invalid_removed:
    total_sum+=age
avg = total_sum/len(invalid_removed)
print(f"Average age of age:: {avg}")
```

Displays the list of customers whose age is greater than 30.

```python
for age in invalid_removed:
    if age > 30:
        print(age, end=" ")
```

```python
print(f"Total {len(ages) + invalid_count} from list has {invalid_count} invalid item")
```
