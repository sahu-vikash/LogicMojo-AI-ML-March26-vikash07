Identify and display all transaction values that appear more than once

```python
transactions = [1200, 3000, 250, 7800, 540, 3000, 9100, 250, 1500, 7800]
duplicate_arr = []
for i in range(len(transactions)):
    for j in range(i+1, len(transactions)):
        if transactions[i] == transactions[j]:
            duplicate_arr.append(transactions[i])

print(f"Duplicate amount of transaction {duplicate_arr}")
```

Detect transactions greater than ₹5000 as these may require fraud monitoring.

```python
transactions = [1200, 3000, 250, 7800, 540, 3000, 9100, 250, 1500, 7800]
for value in transactions:
    if value > 5000:
        transactions.remove(value)

print(f" Unique transactions are :: {transactions}")
```

Calculate the total spending for the day.

```python
transactions = [1200, 3000, 250, 7800, 540, 3000, 9100, 250, 1500, 7800]
spending = 0
for v in transactions:
    spending+=v
print(f" Total spending {spending}")
```

Create a new list containing only unique transaction values.

```python
transactions = [1200, 3000, 250, 7800, 540, 3000, 9100, 250, 1500, 7800]
unique_arr = set()

for value in transactions:
    unique_arr.add(value)
print(f"Unique transactions :: {unique_arr}")
```

Sort the unique transaction list in descending order to view the largest transactions first.

```python
transactions = [1200, 3000, 250, 7800, 540, 3000, 9100, 250, 1500, 7800]
unique_arr = set()

for value in transactions:
    unique_arr.add(value)

sorted_unique = sorted(unique_arr, reverse=True)
print(f"Unique transactions :: {sorted_unique}")
```
