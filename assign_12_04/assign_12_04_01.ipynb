***Understanding Feature Distributions (Iris Dataset)
Context: Before training any ML model, we analyze feature distributions.***

```python
import seaborn  as sns
import pandas as pd

_iris_data = sns.load_dataset("iris")
_iris_data
```

Extract feature matrix as NumPy array

```python
import numpy as np
iris_array_data = np.array(_iris_data)
iris_array_data
```

Compute: Mean Median Standard deviation Variance (for each feature)

```python
numerical_iris_data = _iris_data.select_dtypes(include='number')
numerical_iris_describe = pd.DataFrame({
    'mean': numerical_iris_data.mean(),
    'median': numerical_iris_data.median(),
    'standard deviation': numerical_iris_data.std(),
    'Variance': numerical_iris_data.var()
})
numerical_iris_describe
```

Identify: Which feature has highest variability and why it matters in ML
here, Standard deviation compute variability.

**petal_length feature has highest variability (1.765298) compared to others, as it has higher length variance corresponding to the different species.**<br>
**variability** - ____used in data analysis shows which feature has more information and help to detect outliner___<br>

**higher variability** - ___value changes a lot across sample__<br>

___why it matters in ML?___<br>
**Model learn by mapping, if feature barly change the feature has little to learn
                              if feature is constant cannot predict future output
                              but feature varies, it help to predict output**

why ML models expect this format --

**ML models are based on linear algebra and expect a feature matirx. y = Xw+b <br>
X= feature matrix<br>
w- weight vector<br>
b- Intercept <br>
row = sample<br>
column = feature<br>
if we pass matrix with (1,) or (150,) its a 1D array and model get confused furthermore couldn't able to identify the diredtion. <br>
suppose if we have (150,1) it's 150 sample with 1 feature but with (150,)<br>
for 1D case 150 sample with 1 feature or 1 feature with 150 sample and end up in error as model can not define feature-weight relationship.**
