***Feature Selection using Statistical Understanding<br>
Context: Not all features improve model performance.<br>
Dataset: Iris***

**Compute correlation matrix**

```python
import numpy as np
import seaborn as sns

iris_df = sns.load_dataset('iris')
iris_df
```

```python
iris_df['species'].value_counts()
```

```python
#iris_df['species'] = iris_df['species'].map({'setosa': 0,'versicolor':1,'virginica': 2})
#iris_df
```

```python
correlation_mat = iris_df.corr(numeric_only=True)
correlation_mat
```

**Identify highly correlated features**

```python
correlation_mat.shape
```

```python
upper_triu = correlation_mat.where(np.triu(np.ones(correlation_mat.shape), k=1).astype(bool))
high_corr_pairs = upper_triu.stack()
high_corr_pairs = high_corr_pairs[high_corr_pairs.abs() > 0.8].sort_values(ascending=False)
high_corr_pairs
```

**Drop redundant features**

```python
to_drop = []
correlation_mat = iris_df.corr(numeric_only = True).abs()
upper_triu = correlation_mat.where(np.triu(np.ones(correlation_mat.shape), k=1).astype(bool))

for column in upper_triu.columns:
    if any(upper_triu[column] > 0.85):
        to_drop.append(column)

print(to_drop)
iris_df.drop(columns=to_drop, inplace=True)
iris_df
```
