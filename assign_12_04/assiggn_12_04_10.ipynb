**Create two arrays of shape (3,2)**

```python
import numpy as np
data = [[3,4],
       [3,5],
       [7,1]]
numpy_array = np.array(data)
numpy_array.ndim
```

```python
numpy_array.shape
```

**Perform: Vertical stacking Horizontal stacking**

```python
hstack = np.hstack(numpy_array)
```

```python
np.vstack(numpy_array)
```

**Reshape into (2,6)**

```python
reshape_array_2_6 = np.resize(numpy_array, (2,6))
reshape_array_2_6
```

```python
reshape_array_2_6.shape
```
