***End-to-End ML Preprocessing Pipeline (Titanic)<br>
Context: Prepare dataset for ML model training***

**Load dataset**

```python
import seaborn as sns
import pandas as pd
import numpy as np

titanic_df = sns.load_dataset("titanic")
titanic_df
```

**Handle missing values**

```python
titanic_df.info()
```

As shown in info **age** has **177** and **embark_town**, **embarked** has few missing value.<br>
that needs to be handled.

```python
titanic_df['age'] = titanic_df['age'].fillna(titanic_df['age'].median())
titanic_df
```

```python
age_null_value_count = titanic_df['age'].isna().sum()
print(f"titanic age null value count {age_null_value_count}")
```

```python
titanic_df.dropna(subset=['embark_town'], inplace=True)
titanic_df
```

```python
embarked_null_value_count = titanic_df['embark_town'].isna().sum()
print(f"titanic age null value count {embarked_null_value_count}")
```

```python
titanic_df.drop(columns=['deck'], inplace=True)
titanic_df
```

**Encode categorical variables (sex, embarked)**

```python
titanic_df['sex'] = titanic_df['sex'].map({'male': 0, 'female': 1})
titanic_df
```

**Getting embarked column available unique categorical data that will be used to encode with their corresponding numerical data**

```python
titanic_df['embarked'].value_counts()
```

```python
titanic_df['embarked'] = titanic_df['embarked'].map({'S': 1,'C': 1 ,'Q': 3})
titanic_df
```

**Normalize numerical features<br>**
**Here we have 2 feature that can be normalize that is Age and fare**

```python
titanic_df['age']  = (titanic_df['age'] - titanic_df['age'].min())/(titanic_df['age'].max() - titanic_df['age'].min())
titanic_df['fare']  = (titanic_df['fare'] - titanic_df['fare'].min())/(titanic_df['fare'].max() - titanic_df['fare'].min())
titanic_df
```

**Convert to NumPy array<br>**
**converted only the feature that has numerical data**

```python
features = ['survived','pclass','sex','age','sibsp','parch','fare','embarked']
titanic_df_array = titanic_df[features].to_numpy()
titanic_df_array[0]
```
