***Custom Feature Transformation using apply()<br>
Context: Real ML requires custom logic transformation.<br>
Dataset: Titanic***

**Create age_group: Child / Adult / Senior**

```python
import numpy as np
import seaborn as sns
import pandas as pd

titanic_df = sns.load_dataset("titanic")
titanic_df
```

```python
def get_age_group(age):
    if age < 12:
        return 'Child'
    elif age <= 30:
        return 'Adult'
    elif age > 30:
        return 'Senior'

titanic_df['age_group'] = titanic_df['age'].apply(get_age_group)
titanic_df
```

**Compute survival rate per group**

```python
survival_age_group = titanic_df.groupby('age_group')['survived'].mean()
print(f"Survival of age group adult {survival_age_group.loc['Adult']:.2%}")
print(f"Survival of age group Child {survival_age_group.loc['Child']:.2%}")
print(f"Survival of age group Senior {survival_age_group.loc['Senior']:.2%}")
```

**Interpret: Which segment has highest survival likelihood**

```python
highest_survival =titanic_df.groupby('age_group')['survived'].mean().sort_values(ascending=False)
highest_survival[0:1]
```
