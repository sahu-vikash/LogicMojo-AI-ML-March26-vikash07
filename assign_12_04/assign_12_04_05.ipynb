***Filtering Data for Business Logic<br>***

**Context: Real ML problems require filtering relevant populations.<br>**

***Dataset: Titanic***

**Filter: Female passengers in 1st class**

```python
import seaborn as sns
import pandas as pd
import numpy as np

titanic_df = sns.load_dataset("titanic")
titanic_female_1st_class_df = titanic_df.loc[(titanic_df['sex'] == 'female') & (titanic_df['pclass'] == 1)]
titanic_female_1st_class_df
```

**Compute: Survival rate for this group**

```python
_female_survival_rate = titanic_female_1st_class_df['survived'].mean()
print(f"female Survival Rate: {_female_survival_rate:.2%}")
```

**Compare with overall survival rate**

```python
overall_rate = titanic_df['survived'].mean()
print(f"All Survival Rate: {overall_rate:.2%}")

print(f"compare: {_female_survival_rate - overall_rate:.2%}")
```
