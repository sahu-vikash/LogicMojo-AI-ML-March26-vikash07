***Group-Based Insights (Customer Segmentation Logic)<br>
Context: ML models often depend on aggregated patterns.<br>
Dataset: Tips***

**Compute: Average tip by day Total bill by gender**

```python
import seaborn as sns
import numpy as np
import pandas as pd

tips_df = sns.load_dataset("tips")
tips_df
```

**Compute: Average tip by day Total bill by gender**

```python
avg_tip = tips_df.groupby('day', observed=False)['tip'].mean()
print(f"Average tip by day {avg_tip}")

total_bill = tips_df.groupby('sex',observed=False)['total_bill'].sum()
print(f"Average tip by day {total_bill}")
```

**Create new feature: tip_percentage**

```python
tips_df['tip_percentage'] = (tips_df['tip']/tips_df['total_bill'])*100
tips_df
```

**dentify: Which day has highest tipping behavior**

```python
tip_by_day = tips_df.groupby('day',observed=False)[['tip','tip_percentage']].mean()
result = tip_by_day['tip_percentage'].idxmax()
max_tip = tip_by_day['tip_percentage'].max()
print(f" On {result} has highest tip {max_tip}")
```
