***Advanced Filtering + Subsetting<br>
Context: Training data is often filtered based on business rules.<br>
Dataset: Tips**

**Filter: total_bill > 20 AND tip < 3**
**Select only relevant columns**

```python
import numpy as np
import seaborn as sns
import pandas as pd

tips_df = sns.load_dataset('tips')
total_bill_gt_20_and_tip_gt_3 = tips_df.loc[(tips_df['total_bill'] > 20) &(tips_df['tip']<3)]
total_bill_gt_20_and_tip_gt_3[['total_bill','tip']]
```

**Analyze: Is this segment under-tipping?**

```python
tips_df['tip_percentage'] = (tips_df['tip'] / tips_df['total_bill']) * 100

overall_avg = tips_df['tip_percentage'].mean()

segment_avg = tips_df.groupby('day', observed=False)['tip_percentage'].mean()

under_tippers = segment_avg[segment_avg < overall_avg]
print(f"Overall Average Tip %: {overall_avg:.2f}%")
print("\nSegments Under-Tipping:")
print(under_tippers)
```
