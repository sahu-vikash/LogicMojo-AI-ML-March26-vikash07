***Dataset: Iris***

```python
import seaborn as sns

_iris_df = sns.load_dataset("iris")
_iris_df
```

___Extract: First 100 samples Only last 2 features___

```python
first_100_feature_df = _iris_df.loc[:,['petal_width', 'species']].head(100)
first_100_feature_df
```

***Use boolean masking: Select samples where petal length is greater than dataset mean***

```python
result  = _iris_df.loc[_iris_df['petal_length'] > _iris_df['petal_length'].mean()]
result
```

**Count selected samples**

```python
result.count()
```
