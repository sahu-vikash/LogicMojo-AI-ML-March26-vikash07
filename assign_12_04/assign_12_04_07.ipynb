***Feature Engineering for Model Improvement<br>
Context: Raw data is rarely useful—features must be created.<br>
Dataset: Titanic***

**Create: family_size = sibsp + parch + 1**

```python
import seaborn as sns
import numpy as np
import pandas as pd

titanic_df = sns.load_dataset("titanic")
titanic_df['family_size'] = titanic_df['sibsp'] + titanic_df['parch'] + 1
titanic_df
```

**Create: is_alone (binary feature)**

```python
titanic_df['is_alone'] = (titanic_df['family_size'] == 1).astype(int)
titanic_df
```

Compute survival rate by is_alone

```python
alone_survival_rate = titanic_df.groupby('is_alone')['survived'].mean()
alone_survival_rate
print(f"{alone_survival_rate[0] :.2%}")
print(f"{alone_survival_rate[1] :.2%}")
```
